```python
#extract the months from orderdate to find the monthly trend
#sd_copy=sd_copy.reset_index() #if you peformed any group by function before then you need to reset the index
sd_copy["Month"] = sd_copy["OrderDate"].dt.month
sd_copy
```

```python
#sorting values by Orderdate & setting the index to orderdate
sd_copy.sort_values(by = ['OrderDate'], inplace = True)
sd_copy.set_index('OrderDate', inplace = True) #setting the index to orderdate
sd_copy
```

```python
#grouping by year to find the yearwise sales trend
total_yearsales=(sd_copy.groupby(['order_year']).sum()['total_sales']).sort_values(ascending=True)
total_yearsales=pd.DataFrame(total_yearsales) #creating a pandas dataframe for the output
total_yearsales.reset_index(inplace=True)
total_yearsales
```

```python
plt.figure(figsize = (15,5))
plt.title('Year Wise Sales Trend (2018 - 2020)', fontsize = 18)
plt.bar(total_yearsales['order_year'], total_yearsales['total_sales'], color = '#37C6AB', edgecolor = 'black')
plt.xlabel('order_year', fontsize = 15)
plt.ylabel('total_sales', fontsize = 15)
plt.show()
```

```python
# top 10 states by total sales value
total_statesales=(sd_copy.groupby(['state']).sum()['total_sales']).sort_values(ascending=False).head(10)
total_statesales
```

```python
# Creating a pandas data frame and plotting a bar chart
total_statesales = pd.DataFrame(total_statesales)
total_statesales.reset_index (inplace=True)

plt.figure(figsize = (15,5))
plt.title('Top 10 States by Total Sales Revenue (2018 - 2020)', fontsize = 18)
plt.bar(total_statesales['state'], total_statesales['total_sales'], color = '#37C6AB', edgecolor = 'black', linewidth = 1)
plt.xlabel("STATE", fontsize = 15)
plt.ylabel("SALES", fontsize = 15)
plt.xticks(fontsize = 12, rotation = 90)
plt.yticks(fontsize = 12)
plt.show()
```

```python
# top 10 states by total sales value
total_stateqty=(sd_copy.groupby(['state']).sum()['Order Quantity']).sort_values(ascending=False).head(10)
total_stateqty=pd.DataFrame(total_stateqty)
total_stateqty.reset_index(inplace = True)
total_stateqty
```

```python
plt.figure(figsize = (15,5))
plt.title('Top 10 States by Total Order Quantity (2018 - 2020)', fontsize = 18)
plt.bar(total_stateqty['state'], total_stateqty['Order Quantity'], color = '#37C6AB', edgecolor = 'black', linewidth = 0.5)
plt.xlabel("STATE", fontsize = 15)
plt.ylabel("ORDER QUANTITY", fontsize = 15)
plt.xticks(fontsize = 12, rotation = 90)
plt.yticks(fontsize = 12)
plt.show()
```

```python
# top 10 products by total Sales 
total_product=(sd_copy.groupby(['product_name'])).sum()['total_sales'].sort_values(ascending=False).head(10)
total_product=pd.DataFrame(total_product)
total_product.reset_index(inplace = True)
total_product
```

```python
total_revenue_product = total_product['total_sales'].sum()
total_revenue_product = str(int(total_revenue_product))
total_revenue_product = '$' + total_revenue_product
total_revenue_product
```

```python
# Visualization of products by total sales

plt.rcParams['figure.figsize'] = (13,7)
plt.rcParams['font.size'] = 12.0
plt.rcParams['font.weight'] = 9
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return ' ${v:d}'.format(v = val)
    return my_format
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#55B4B0','#E15D44','#009B77','r','g']
explode = (0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05)
fig1, ax1 = plt.subplots()
pie1 = ax1.pie(total_product['total_sales'], colors = colors, labels = total_product['product_name'], autopct = autopct_format(total_product['total_sales']), startangle = 90, explode = explode)
fraction_text_list = pie1[2]
for text in fraction_text_list:
    text.set_rotation(315)
center_circle = plt.Circle((0,0), 0.80, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
ax1.axis('equal')
label = ax1.annotate('Total Revenue \n' + str(total_revenue_product), color = 'b', xy = (0,0), fontsize = 15, ha  ='center')
plt.tight_layout()
plt.show()
```

```python
# top 10 Products by Quantity
total_qtyproduct=(sd_copy.groupby(['product_name'])).sum()['Order Quantity'].sort_values(ascending=False).head(10)
total_qtyproduct=pd.DataFrame(total_qtyproduct)
total_qtyproduct.reset_index (inplace = True)
total_qtyproduct
```

```python
plt.figure(figsize = (15,5))
plt.title('Top 10 Products by Total Quantity (2018 - 2020)', fontsize = 18)
plt.bar(total_qtyproduct['product_name'], total_qtyproduct['Order Quantity'], color = '#37C6AB', edgecolor = 'black', linewidth = 0.8)
plt.xlabel("Product", fontsize = 15)
plt.ylabel("Quantity", fontsize = 15)
plt.xticks(fontsize = 12, rotation = 90)
plt.yticks(fontsize = 12)
plt.show()
```
